```python
### Assignment 2: Logging with Different Handlers

#1. Write a Python function to create a logger that logs messages to both a file named `app.log` and the console.
#2. Modify the function to use different logging levels for the file and console handlers.

import logging

# Configure basic logging settings
logging.basicConfig(
    level=logging.DEBUG,  # Set the overall logging level
    format='%(asctime)s - %(name)s - %(levelname)s: %(message)s',  # Log message format
    datefmt='%Y-%m-%d %H:%M:%S'  # Date format
)

# Create a logger object
logger = logging.getLogger("MyLogger")

# Create a file handler
file_handler = logging.FileHandler('applog.log')  # Log messages to 'applog.log'
file_handler.setLevel(logging.INFO)  # Set file handler level to INFO

# Create a console handler
console_handler = logging.StreamHandler()  # Log messages to the console
console_handler.setLevel(logging.DEBUG)  # Set console handler level to DEBUG

# Create a formatter and set it for both handlers
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s: %(message)s', datefmt='%Y-%m-%d %H:%M:%S')
file_handler.setFormatter(formatter)
console_handler.setFormatter(formatter)

# Add handlers to the logger
logger.addHandler(file_handler)
logger.addHandler(console_handler)

# Example usage of the logger
logger.debug("This is a debug message")
logger.info("This is an info message")
logger.warning("This is a warning message")
logger.error("This is an error message")
logger.critical("This is a critical message")
```

```python
### Assignment 6: Contextual Logging

#1. Write a Python function to create a logger that includes contextual information (e.g., function name, line number) in the log messages.
#2. Modify the function to include additional contextual information (e.g., user ID, session ID).

#1
import logging

def setup_logger_with_context(log_file='contextual.log'):
    """
    Set up a logger that logs messages with contextual information like function name and line number.
    
    :param log_file: Name of the log file to store logs.
    :return: Configured logger
    """
    logger = logging.getLogger("ContextLogger")
    logger.setLevel(logging.DEBUG)

    # Create file handler
    file_handler = logging.FileHandler(log_file)
    file_handler.setLevel(logging.DEBUG)

    # Define the log format to include function name and line number
    formatter = logging.Formatter(
        '%(asctime)s - %(name)s - %(levelname)s - %(message)s - [Func: %(funcName)s | Line: %(lineno)d]',
        datefmt='%Y-%m-%d %H:%M:%S'
    )
    
    file_handler.setFormatter(formatter)
    logger.addHandler(file_handler)

    return logger

# Example usage
def example_function():
    logger = setup_logger_with_context()
    logger.debug("This is a debug message.")
    logger.info("This is an info message.")

example_function()


#2
import logging

class ContextFilter(logging.Filter):
    """
    A custom filter to add user-defined contextual information to the log messages.
    """
    def __init__(self, user_id=None, session_id=None):
        super().__init__()
        self.user_id = user_id
        self.session_id = session_id

    def filter(self, record):
        # Add custom attributes to the log record
        record.user_id = self.user_id if self.user_id else 'N/A'
        record.session_id = self.session_id if self.session_id else 'N/A'
        return True

def setup_logger_with_additional_context(log_file='contextual_extended.log', user_id=None, session_id=None):
    """
    Set up a logger that logs messages with additional contextual information like user ID and session ID.
    
    :param log_file: Name of the log file to store logs.
    :param user_id: Custom user ID to include in the logs.
    :param session_id: Custom session ID to include in the logs.
    :return: Configured logger
    """
    logger = logging.getLogger("ExtendedContextLogger")
    logger.setLevel(logging.DEBUG)

    # Create file handler
    file_handler = logging.FileHandler(log_file)
    file_handler.setLevel(logging.DEBUG)

    # Define the log format to include function name, line number, user ID, and session ID
    formatter = logging.Formatter(
        '%(asctime)s - %(name)s - %(levelname)s - %(message)s - [Func: %(funcName)s | Line: %(lineno)d | '
        'User: %(user_id)s | Session: %(session_id)s]',
        datefmt='%Y-%m-%d %H:%M:%S'
    )
    
    file_handler.setFormatter(formatter)

    # Add the custom filter to inject user ID and session ID into logs
    context_filter = ContextFilter(user_id=user_id, session_id=session_id)
    logger.addFilter(context_filter)

    logger.addHandler(file_handler)

    return logger

# Example usage
def example_function_with_context():
    logger = setup_logger_with_additional_context(user_id="User123", session_id="Session456")
    logger.debug("This is a debug message with context.")
    logger.info("This is an info message with context.")

example_function_with_context()
```

```python
### Assignment 7: Configuring Logging with a Dictionary

#1. Write a Python function to configure logging using a dictionary. The configuration should include handlers for both file and console logging.
#2. Modify the dictionary to include different logging levels and formats for each handler.

import logging
import logging.config

def setup_logging():
    # Dictionary configuration for logging
    log_config = {
        'version': 1,  # Logging configuration version
        'disable_existing_loggers': False,  # Keep the existing loggers intact
        'formatters': {
            'detailed': {
                'format': '%(asctime)s - %(name)s - %(levelname)s - %(message)s'
            },
            'simple': {
                'format': '%(levelname)s - %(message)s'
            }
        },
        'handlers': {
            'console': {
                'class': 'logging.StreamHandler',  # Output to the console
                'level': 'DEBUG',  # Console handler log level
                'formatter': 'simple',  # Use the simple formatter
                'stream': 'ext://sys.stdout'  # Default is stdout
            },
            'file': {
                'class': 'logging.FileHandler',  # Output to a file
                'level': 'DEBUG',  # File handler log level
                'formatter': 'detailed',  # Use the detailed formatter
                'filename': 'app.log',  # Log file name
                'mode': 'a'  # Append mode
            }
        },
        'root': {
            'level': 'DEBUG',  # Overall logging level
            'handlers': ['console', 'file']  # Both console and file handlers
        }
    }

    # Apply the logging configuration
    logging.config.dictConfig(log_config)

# Test the logging configuration
def test_logging():
    setup_logging()

    logger = logging.getLogger(__name__)  # Create a logger
    logger.debug("This is a debug message")
    logger.info("This is an info message")
    logger.warning("This is a warning message")
    logger.error("This is an error message")
    logger.critical("This is a critical message")

# Run the test function
test_logging()
```
