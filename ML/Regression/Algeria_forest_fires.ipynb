```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
#datacleaning 
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
#added the new region columns
dataset.loc[:122,"Region"]=0
```

```python
dataset.loc[122:,"Region"]=1
```

```python
dataset.info()
#will sees the data type of region as float
```

```python
dataset[['Region']]=dataset[['Region']].astype(int)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
df=dataset.dropna().reset_index(drop=True)
#it will adjustes the index values 
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
#will see the column names having the white spaces
```

```python
#fixing the white space in column name s
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.columns
```

```python
#change the data type of to columns to integer as required 
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
#changnht the other coumsns to float values 
objects=[feature for feature in df.columns if df[feature].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
        
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv("Agerian_forest_fires_cleaned_datasets.csv")
```

```python
#exploratroy analysis 
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
#encoding the categories in classes 
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
print(plt.style.available)
```

```python
#plot the density plot for all features 
import matplotlib.pyplot as plt
import seaborn
#plt.style.use('seaborn')
plt.style.use("seaborn-v0_8-dark-palette")
```

```python
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
#percentage for pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
#plotting the pie chart
classlabels=['fire','not fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct="%1.1f")
plt.title("Pie chart of classes")
plt.show()
```
