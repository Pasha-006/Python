```python
x=df.drop('selling_price',axis=1)
y=df['selling_price']
```

```python
#train split the data 
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

Using Label encoders to encoding the categorical columns

```python
from sklearn.preprocessing import LabelEncoder 
LabelEncoder1=LabelEncoder()
LabelEncoder2=LabelEncoder()
LabelEncoder3=LabelEncoder()
LabelEncoder4=LabelEncoder()
LabelEncoder5=LabelEncoder()
x_train['brand']=LabelEncoder1.fit_transform(x_train['brand'])
x_train['model']=LabelEncoder2.fit_transform(x_train['model'])
x_train['seller_type']=LabelEncoder3.fit_transform(x_train['seller_type'])
x_train['transmission_type']=LabelEncoder4.fit_transform(x_train['transmission_type'])
x_train['fuel_type']=LabelEncoder5.fit_transform(x_train['fuel_type'])
```

```python
x_train
```

```python
x_test['brand']=LabelEncoder1.transform(x_test['brand'])
x_test['model']=LabelEncoder2.transform(x_test['model'])
x_test['seller_type']=LabelEncoder3.transform(x_test['seller_type'])
x_test['transmission_type']=LabelEncoder4.transform(x_test['transmission_type'])
x_test['fuel_type']=LabelEncoder5.transform(x_test['fuel_type'])
```

```python
#model Training 
from sklearn.linear_model import LinearRegression 
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```
