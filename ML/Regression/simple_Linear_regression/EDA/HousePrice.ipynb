```python
def cost_function(w,b): 
    n=len(x)
    error=0
    for i in range(0,n):
        error+=(w*x2[i]+b-y[i])**2 
    
    return ((1/(2*n))*error) 

    
```

```python
def gradient_descent(w,b,alpha):
    w_update=0 
    b_update=0
    n=len(x)
    for i in range(0,len(x2)):
        w_update+=(w*x2[i]+b-y[i])*x2[i]
        b_update+=(w*x2[i]+b-y[i]) 
    w=w-alpha*(w_update/n)
    b=b-alpha*(b_update/n)
    print(w,b)
    return w,b
```

```python
def training(epochs,alpha,w,b):
    cost_history=[]
    gradient_hist=[]
    for i in range(epochs):
        gradient=gradient_descent(w,b,alpha)
        print(gradient)
        cost=cost_function(w,b)
        cost_history.append(gradient)
        gradient_hist.append(cost)
    return cost,gradient,cost_history

```

```python
cost,gradient,cost_history=training(200,0.5,1,1) 
```

```python
print(cost)
print(gradient)
print(cost_history)
```
