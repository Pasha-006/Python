```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##sentences 
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
     'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
##define the vocabulary size
voc_size=10000
```

```python
##one hot representation  
one_hot_repr=[one_hot(words,voc_size)for words in sent]
```

```python
one_hot_repr
```

```python
#word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences 
from tensorflow.keras.models import Sequential 
```

```python
import numpy as np 
```

```python
#its importnant that all sentences should have equal word lenght becuase our rnn size should be able to perform actions or time stamps 
```

```python
sent_length=8
#making all words in sentence to 8 if its have less than 8 we are adding the empty sentence to it 
embedded_doc=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_doc)
```

```python
### feature represemtation
dim=10 
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim
                    ,input_length=sent_length))
model.compile('adam','mse')
#embedding internallly uses word2vec aporach to conver int the vectors
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
