```python
import pandas as pd

# Initialize empty lists to store labels and messages
labels = []
messages = []

# Open the file and read line by line
with open("spam.csv", encoding='latin1') as file:
    for line in file:
        # Split the line at the first comma
        parts = line.split(',', 1)
        # Append the label and message to the respective lists
        labels.append(parts[0])
        messages.append(parts[1].strip())

# Create a DataFrame from the lists
messages = pd.DataFrame({'label': labels, 'message': messages})

# Display the first few rows to check if it's read correctly
print(messages.head())
```

```python
messages['label']
```

```python
messages['label']=messages['label'].replace('ham"""','ham')
```

```python
import re 
import nltk
import nltk.downloader 
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords 
from nltk.stem.porter import PorterStemmer 
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
    
```

```python
corpus
```

## Create a Bag of words


```python
from sklearn.feature_extraction.text import CountVectorizer 
##for binary Bow enable binary=True 
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
#Independent feature values
x=cv.fit_transform(corpus).toarray()
```

```python
import numpy as np 
np.set_printoptions(edgeitems=30,linewidth=100000,
                    formatter=dict(float = lambda x : "%.3g" %x))
x
```

```python
cv.vocabulary_
```

```python
#outut features 
y=pd.get_dummies(messages['label'])
```

```python
y=y.astype(int)
```

```python
y
```

```python
#dependent values 
y=y['spam'] #getting the all values of ham class as expected outocme
```

```python
y.shape
```

```python
#train test split 
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
#naive bayes works well witht the sparse matrix kind of data 
from sklearn.naive_bayes import MultinomialNB 
```

```python
spam_detect_model=MultinomialNB().fit(x_train,y_train)
```

```python
y_pred=spam_detect_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report 
print(classification_report(y_test,y_pred))
```

###TfIdf call 


```python

from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
x=tv.fit_transform(corpus).toarray()
```

```python
tv.vocabulary_
```

```python
#train test split 
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model=MultinomialNB().fit(x_train,y_train)
```

```python
#prediction 
y_pred=spam_tfidf_model.predict(x_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report 
print(classification_report(y_pred,y_test))
```
