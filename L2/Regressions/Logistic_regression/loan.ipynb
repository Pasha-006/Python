```python
#splitting the data set 
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=45)
```

```python
x_train
```

```python
#Label encoding on person gender 
from sklearn.preprocessing import LabelEncoder 
label1=LabelEncoder() 
lable2=LabelEncoder() 
label3=LabelEncoder()
label4=LabelEncoder()
label5=LabelEncoder()
x_train['person_gender']=label1.fit_transform(x_train['person_gender'])
x_train['person_education']=lable2.fit_transform(x_train['person_education'])
x_train['person_home_ownership']=label3.fit_transform(x_train['person_home_ownership'])
x_train['loan_intent']=label4.fit_transform(x_train['loan_intent'])
x_train['previous_loan_defaults_on_file']=label5.fit_transform(x_train['previous_loan_defaults_on_file'])
```

```python
import pickle
encoders=[label1,lable2,label3,label4,label5]
pickle.dump(encoders,open("labelencoders.pkl","wb"))

```

```python
x_test['person_gender']=label1.transform(x_test['person_gender']) 
x_test['person_education']=lable2.transform(x_test['person_education'])
x_test['person_home_ownership']=label3.transform(x_test['person_home_ownership'])
x_test['loan_intent']=label4.transform(x_test['loan_intent'])
x_test['previous_loan_defaults_on_file']=label5.transform(x_test['previous_loan_defaults_on_file'])
```

```python
#Logistic regression
```

```python
from sklearn.linear_model import LogisticRegression 
logistic_model=LogisticRegression() 
logistic_model.fit(x_train,y_train)
y_pred=logistic_model.predict(x_test) 
```

```python
coeff=logistic_model.predict([x_test.loc[23491]])
print(coeff)
```

```python
coeff
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error 
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rsme=np.sqrt(mse)
print(mse)
print(mae)
print(rsme)
```

```python
#model score 
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score =accuracy_score(y_test,y_pred)
print(score)
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
report=classification_report(y_test,y_pred)
print(report)
```

```python
#hyper parameter tuning 
model=LogisticRegression() 
penalty=['l1','l2','elasticnet']
c_values=[100,10,1,0.1,0.01]
solver=['lbfgs','liblinear', 'newton-cg', 'newton-cholesky', 'sag','saga'] 
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
#grdi search cv 
from sklearn.model_selection import GridSearchCV 
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
grid.best_params_
```

```python
y_pred=grid.predict_proba(x_test)
```
